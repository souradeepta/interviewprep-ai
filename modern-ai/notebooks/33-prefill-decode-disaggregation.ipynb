```
import numpy as np; import torch; device = torch.device("cuda" if torch.cuda.is_available() else "cpu"); print(f"Device: {device}")
```

## Level 1: Basic

```
X = np.random.randn(4, 8, 16); print(f"Shape: {X.shape}"); Y = X * 1.0; print(f"Output: {Y.shape}")
```

## Level 2: Advanced

```
import torch.nn as nn
model = nn.Linear(16, 16)
x = torch.randn(4, 8, 16)
with torch.no_grad():
    y = model(x)
print(f"Output: {y.shape}")
```

## Example 1

```
model = nn.Linear(64, 64)
x = torch.randn(8, 64)
y = model(x)
print(f"Result: {y.shape}")
```

## Example 2

```
try:
    model = nn.Linear(256, 256)
    x = torch.randn(32, 256)
    y = model(x)
    print("Success")
except Exception as e:
    print(f"Error: {e}")
```

## Example 3

```
shapes = [(1, 16), (8, 64), (32, 256)]
for s in shapes:
    x = torch.randn(*s)
    y = torch.relu(x)
    print(f"✓ {s}")
```
