# Linear Regression From Scratch

**What you'll build:** Normal equation (closed-form) + gradient descent implementations.

**Prerequisites:** NumPy basics, matrix multiplication, calculus intuition for derivatives.

## Concept Recap

**Goal:** find weights $\theta$ minimizing MSE: $\mathcal{L} = \frac{1}{n}\sum (y_i - \hat{y}_i)^2$

**Normal Equation (closed-form):** $\theta = (X^TX)^{-1}X^Ty$
- Exact solution. O(d³) — slow when features d is large.

**Gradient Descent:** $\theta \leftarrow \theta - \eta \nabla_\theta \mathcal{L}$
- Gradient: $\nabla_\theta \mathcal{L} = \frac{2}{n} X^T(X\theta - y)$
- Iterative. O(nd) per step. Scales to large datasets.

**Bias term:** add column of 1s to X (X_b) so intercept is part of $\theta$.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
X = np.random.randn(100, 1)
y = 3 * X.squeeze() + 2 + np.random.randn(100) * 0.5

X_b = np.hstack([np.ones((100, 1)), X])  # add bias column

# Normal equation
w_ne = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
print(f"Normal equation: intercept={w_ne[0]:.3f}, slope={w_ne[1]:.3f}")

# Gradient descent
w = np.zeros(2)
lr = 0.1
losses = []
for _ in range(1000):
    grad = (2 / 100) * X_b.T @ (X_b @ w - y)
    w -= lr * grad
    losses.append(np.mean((X_b @ w - y) ** 2))

print(f"Gradient descent: intercept={w[0]:.3f}, slope={w[1]:.3f}")
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(X, y)
print(f"sklearn: intercept={model.intercept_:.3f}, slope={model.coef_[0]:.3f}")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

x_line = np.linspace(-3, 3, 100)
ax1.scatter(X, y, alpha=0.5, label='data')
ax1.plot(x_line, w_ne[0] + w_ne[1]*x_line, 'r-', label='normal eq')
ax1.plot(x_line, w[0] + w[1]*x_line, 'g--', label='gradient descent')
ax1.set_xlabel('X'); ax1.set_ylabel('y'); ax1.legend(); ax1.set_title('Fitted Lines')

ax2.plot(losses)
ax2.set_xlabel('Iteration'); ax2.set_ylabel('MSE Loss'); ax2.set_title('Training Loss Curve')

plt.tight_layout()
plt.show()
```

## Summary

- Normal equation gives exact solution in one step — use when n and d are small
- Gradient descent scales to large n and d — standard for neural networks
- Both converge to the same solution on convex MSE loss
